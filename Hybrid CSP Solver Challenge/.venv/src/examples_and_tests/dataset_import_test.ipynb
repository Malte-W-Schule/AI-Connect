Test to load the top 100 datasets via given api

```python
import requests

def fetch_rows(
    dataset="allenai/ZebraLogicBench",
    config="grid_mode",
    split="test",
    offset=0,
    length=2,
):
    url = (
        "https://datasets-server.huggingface.co/rows"
        "?dataset={}&config={}&split={}&offset={}&length={}"
    ).format(
        requests.utils.quote(dataset, safe=""),
        config,
        split,
        offset,
        length
    )

    response = requests.get(url)
    response.raise_for_status()
    return response.json()

if __name__ == "__main__":
    data = fetch_rows()
    print(data)
```
