Defining the constrains as lambda:

```python
nextToCon =     lambda x,y : abs(x-y) == 1

notNextToCon =  lambda x,y : abs(x-y) != 1

sameCon =       lambda x,y : x == y

#...
```

gives us the option to dynamically use them later withouth hardcoding any

```python
def useConstraint(nodeleft,noderight,constraint):
    print(f"result:{constraint(nodeleft,noderight)}")
```
